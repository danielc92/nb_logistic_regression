```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cancer = load_breast_cancer()
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify = cancer.target, random_state = 42)

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
print("Accuracy on training subset: {:.3f}".format(log_reg.score(X_train, y_train)))
print("Accuracy on test subset: {:.3f}".format(log_reg.score(X_test, y_test)))
```

```python
log_reg100 = LogisticRegression(C=100)
log_reg100.fit(X_train, y_train)
print("Accuracy on training subset: {:.3f}".format(log_reg100.score(X_train, y_train)))
print("Accuracy on test subset: {:.3f}".format(log_reg100.score(X_test, y_test)))
```

```python
log_reg001 = LogisticRegression(C=.01)
log_reg001.fit(X_train, y_train)
print("Accuracy on training subset: {:.3f}".format(log_reg001.score(X_train, y_train)))
print("Accuracy on test subset: {:.3f}".format(log_reg001.score(X_test, y_test)))
```
